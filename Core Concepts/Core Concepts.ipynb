# 缠论
## 1. 基本定义
### 1.1 缺口
**缺口**: 同级别K线图上两相邻K线间出现没有成交的区间

**缺口回补**：该缺口区间最终全部再次出现成交的过程

**回补时限**：可能下一K线就回补，也可能永远不出现

**行情力度(根据缺口回补分类):** 

(1)	不回补，强势;  
(2)	回补后继续新高或新低，平势;  
(3)	回补后不能新高、新低，因而出现原来走势的转折，弱势。

**缺口级别：**

日K线图有缺口，则日线以下的任何周期的K线图都会相应有缺口；而回补日线的缺口，不一定能回补日线以下周期K线图上的缺口。

**缺口性质：**

盘整走势中的缺口属于普通缺口，一般都回补，分析意义不大，唯一的意义，就是在中枢震荡中有一个目标，就是回拉的过程中，几乎肯定能至少拉回补掉缺口的位置。

**底分型：** 第二K线低点是相邻三K线低点中最低的，而高点也是相邻三K线高点中最低的，底分型的最低点叫该分型的底，图2

### 1.3 笔

两个相邻的顶和底之间构成一笔，即顶和底之间的其他波动都忽略不计

(1)	顶分型与底分型经过包含处理后，不允许共用K线，也就是不能有一K线分别属于顶分型与底分型，这条件和原来是一样的，这样才能保证足够的能量力度；如下图中的a满足，b不满足。
